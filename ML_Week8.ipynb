**Program 10**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Step 1: Import necessary libraries

# Step 2: Load the dataset
data = pd.read_csv("/Iris.csv")

# Step 3: Build the k-Means Model
# Considering 3 clusters for Iris dataset
kmeans = KMeans(n_clusters=3)
kmeans.fit(data[['PetalLengthCm', 'PetalWidthCm']])

# Step 4: Visualize the Clustering Results
# Plot the original classifications using petal features
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=data)
plt.title('Original Classifications using Petal Features')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()

# Plot the model's classifications
data['kmeans_cluster'] = kmeans.labels_
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='kmeans_cluster', data=data, palette='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
plt.title('Model\'s Classifications')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Convert to DataFrame for visualization
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
df_pca['target'] = y

# Plot the PCA results
plt.figure(figsize=(10, 6))
plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df_pca['target'], cmap='viridis', alpha=0.8)
plt.title('PCA of Breast Cancer Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Target')
plt.show()
```
